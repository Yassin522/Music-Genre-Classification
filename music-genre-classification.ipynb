# Data Pre-Processing

```python
train.drop(['id','key'],axis=1,inplace=True)
test.drop(['id','key'],axis=1,inplace=True)
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='median')
train[['Popularity','instrumentalness']]=imp.fit_transform(train[['Popularity','instrumentalness']])
```

```python
test.Popularity = test.Popularity.fillna(value=train.Popularity.median(), axis=0)
test.instrumentalness = test.instrumentalness.fillna(value=train.instrumentalness.median(), axis=0)
```

```python
X=train.drop('Class',axis=1)
y=train['Class']
```

```python
cat_cols = ['Artist Name', 'Track Name']
```

# Train model

```python
from catboost import CatBoostClassifier
cat_clf = CatBoostClassifier(n_estimators=1500, depth =6 ,learning_rate=0.05, random_state=10, eval_metric= 'MultiClass', cat_features=cat_cols)
cat_clf.fit(X,y)
```
